```python
import pandas as pd

# Assuming df is your DataFrame

# Check if 'occupation' column exists before dropping
if 'occupation' in df.columns:
    # One-hot encode 'occupation' column
    df = pd.concat([df.drop('occupation', axis=1), pd.get_dummies(df['occupation']).add_prefix('occupation_')], axis=1)

# Check if 'workclass' column exists before dropping
if 'workclass' in df.columns:
    # One-hot encode 'workclass' column
    df = pd.concat([df.drop('workclass', axis=1), pd.get_dummies(df['workclass']).add_prefix('workclass_')], axis=1)

# Drop 'education' column
df = df.drop('education', axis=1)

# Check if 'marital-status' column exists before dropping
if 'marital-status' in df.columns:
    # One-hot encode 'marital-status' column
    df = pd.concat([df.drop('marital-status', axis=1), pd.get_dummies(df['marital-status']).add_prefix('marital-status_')], axis=1)

# Check if 'relationship' column exists before dropping
if 'relationship' in df.columns:
    # One-hot encode 'relationship' column
    df = pd.concat([df.drop('relationship', axis=1), pd.get_dummies(df['relationship']).add_prefix('relationship_')], axis=1)

# Check if 'race' column exists before dropping
if 'race' in df.columns:
    # One-hot encode 'race' column
    df = pd.concat([df.drop('race', axis=1), pd.get_dummies(df['race']).add_prefix('race_')], axis=1)

# Check if 'native-country' column exists before dropping
if 'native-country' in df.columns:
    # One-hot encode 'native-country' column
    df = pd.concat([df.drop('native-country', axis=1), pd.get_dummies(df['native-country']).add_prefix('native-country_')], axis=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

df = df.drop('fnlwgt', axis =1)

train_df, test_df = train_test_split(df, test_size=0.2)
```

```python
train_df
```

```python
test_df
```

```python
train_X = train_df.drop('income', axis=1)
train_Y = train_df['income']

test_X = test_df.drop('income', axis=1)
test_Y = test_df['income']
```

```python
forest = RandomForestClassifier()
forest.fit(train_X, train_Y)
```

```python
forest.score(test_X, test_Y)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

param_grid = {
    'n_estimators': [50, 100, 250],
    'max_depth': [5, 10, 30, None],  # Corrected the value 'NONE' to 'None'
    'min_samples_split': [2, 4],  # Corrected the parameter name 'min_samples_splits' to 'min_samples_split'
    'max_features': ['sqrt', 'log2']
}

grid_search = GridSearchCV(estimator=RandomForestClassifier(),
                           param_grid=param_grid, verbose=10)
```

```python
grid_search.fit(train_X,train_Y)
```

```python
grid_search.best_estimator_
```

```python
forest = grid_search.best_estimator_
```
